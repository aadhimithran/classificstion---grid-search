```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv('Social_Network_Ads 1.csv')
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset = pd.get_dummies(dataset, columns=['Gender_Male'], dtype=int, drop_first=True)
```

```python
dataset
```

```python
dataset["Purchased"].value_counts()
```

```python
dataset.columns
```

```python
independent=dataset[['User ID', 'Age', 'EstimatedSalary','Gender_Male_1']]
```

```python
independent
```

```python
dependent=dataset[['Purchased']]
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
x_train
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
params_grid={'solver':['newton-cg','lbfgs','liblinear','saga'],'penalty':['l1']}
grid=GridSearchCV(LogisticRegression(),params_grid,refit=True,verbose=3,n_jobs=1,scoring='f1_weighted')
```

```python
grid.fit(x_train,y_train)
```

```python
re=grid.cv_results_
grid_predictions=grid.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,grid_predictions)
```

```python
from sklearn.metrics import classification_report
clf_report=classification_report(y_test,grid_predictions)
```

```python
print(cm)
print(clf_report)
```

```python

from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,grid.predict_proba(x_test)[:,1])
```

```python
table = pd.DataFrame.from_dict(re)
```

```python
table
```

```python
# Grid searchCV using RandomForest
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.ensemble import RandomForestClassifier
params_grid={'criterion':['gini','entropy'],'max_features': ['auto','sqrt','log2'],'n_estimators':[10,100]}
grid = GridSearchCV(RandomForestClassifier(), params_grid, refit=True, verbose=3, n_jobs=-1)
grid.fit(x_train ,y_train)
```

```python
re=grid.cv_results_
grid_predictions=grid.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,grid_predictions)
```

```python
from sklearn.metrics import classification_report
clf_report=classification_report(y_test,grid_predictions)
```

```python
print(cm)
print(clf_report)
```

```python

from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,grid.predict_proba(x_test)[:,1])
```

```python
# k- nearest neighbour
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
```

```python
params_grid = {
    'n_neighbors': [3, 5, 7, 9],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan', 'minkowski']
}
grid = GridSearchCV(KNeighborsClassifier(), params_grid, refit=True, verbose=3, n_jobs=-1, cv=5)
grid.fit(x_train ,y_train)
```

```python
re=grid.cv_results_
grid.predictions=grid.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,grid.predictions)
```

```python
from sklearn.metrics import classification_report
clf_report=classification_report(y_test,grid.predictions)
```

```python
print(cm)
print(clf_report)
```

```python

from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,grid.predict_proba(x_test)[:,1])
```
